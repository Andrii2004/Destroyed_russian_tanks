```python
import torch.cuda
from ultralytics import YOLO
```

```python
print('Cuda is available = ',torch.cuda.is_available())
print('Cuda device count = ',torch.cuda.device_count())
```

```python
model = YOLO("yolo11s.pt")
```

```python
results = model.train(data='/kaggle/input/tanks-images/tanks_all_final.v9i_yolov11/data.yaml', 
                      epochs=300, imgsz=640, optimizer = 'Adam', plots=True , 
                      device=0 ,cache = 'ram' ,project='Yolo_11_s_v3',save_period = 20 ,
                      lr0=0.001,batch = -1,patience= 50)
```

```python
model.export(format="onnx")
```
