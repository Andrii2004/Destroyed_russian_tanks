```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")
```

```python
results = model.train(data='/kaggle/input/tanks-v2/data.yaml', 
                      epochs=300, imgsz=640, optimizer = 'Adam', plots=True , 
                      device=0 ,cache = True ,project='Yolo_11_m_3',save_period = 20 ,
                      lr0=0.001,batch = -1,patience= 50)
```

```python
metrics = model.val()
```
